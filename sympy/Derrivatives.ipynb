# Higher order derrivative

diff can take multiple derivatives at once. To take multiple derivatives, pass the variable as many times as you wish to differentiate, or pass a number after the variable. 

<img src = "https://slideplayer.com/slide/16844340/97/images/5/1.8+Higher+Order+Derivatives.jpg">

```python
x = symbols('x')
```

```python
fx = x ** 5 - 3 * x ** 4 + x
```

```python
display(fx)
```

```python
display(diff(fx, x, 5))
```

```python
display(expand(diff(fx, x, 3)))
```

# Partial derrivative
You can also take derivatives with respect to many variables at once. Just pass each derivative in order, using the same syntax as for single variable derivatives. For example, each of the following will compute  
$$f(x) = e^{x y z}$$

Evaluate:


$$ \frac{\partial^7}{\partial x\partial y^2\partial z^4} f(x)$$

```python
x,y,z = symbols('x y z')
```

```python
fx = exp(x*y*z)
```

```python
display(fx)
```

```python
diff(fx, x,y, y,z,z,z,z)
```

```python
diff(fx, x,1, y, 2, z, 4 )
```

# Optional

Verify that the function $y = e^{-3x}$ is a solution to the differential equation.


$$\begin{array}{l}\frac{d^{2}y}{dx^{2}}+\frac{dy}{dx}  6y = 0\end{array}$$

```python
x,y = symbols('x y')
```

```python
y = exp(-3 * x)
```

```python
out = diff(y,x,2) + diff(y, x, 1) - 6 * y
```

```python
display(out)
```

```python
d2dx2 = diff(y, x, 2)
```

```python
display(d2dx2)
```

```python
ddx = diff(y, x, 1)
```

```python
display(ddx)
```

```python
output = d2dx2 + ddx - 6 * y
```

```python
display(output)
```
